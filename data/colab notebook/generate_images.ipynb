```python
from diffusers import StableDiffusionXLPipeline
import torch

model_id = "stabilityai/stable-diffusion-xl-base-1.0"

pipe = StableDiffusionXLPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True,
    add_watermarker=False  # ðŸ”‘ Prevents forced square images
)

pipe.to("cuda")
print("âœ… SDXL model loaded and ready with custom size support")
```

```python
import json
import os
from tqdm import tqdm
from google.colab import files
import gc

# ðŸ“‚ Set output folder
output_dir = "/content/ai_video_editor/generated_images"
os.makedirs(output_dir, exist_ok=True)

# ðŸ—‚ï¸ Upload image_prompts.json
print("ðŸ“ Please upload your image_prompts.json file")
uploaded = files.upload()

# Use uploaded file name dynamically
prompt_file = list(uploaded.keys())[0]

# â±ï¸ Ask for inference steps
inference_steps = int(input("Enter number of inference steps (e.g. 40): "))

# ðŸ“– Load prompts
with open(prompt_file, "r", encoding="utf-8") as f:
    prompts_data = json.load(f)

# ðŸ–¼ï¸ Generate images
for idx, item in tqdm(enumerate(prompts_data), desc="Generating", total=len(prompts_data)):
    prompt = item["prompt"]
    width = item.get("width", 1024)
    height = item.get("height", 576)

    # ðŸ” Make sure width and height are divisible by 8
    width = (width // 8) * 8
    height = (height // 8) * 8

    print(f"\nðŸŽ¨ Generating image {idx+1}: {prompt}")
    print(f"ðŸ“ Size: {width}x{height}")

    try:
        # ðŸ–Œï¸ Generate image
        image = pipe(
            prompt=prompt,
            num_inference_steps=inference_steps,
            guidance_scale=12.5,
            height=height,
            width=width
        ).images[0]

        # ðŸ’¾ Save image
        image_path = os.path.join(output_dir, f"image_{idx+1:03d}.png")
        image.save(image_path)
        print(f"âœ… Saved: {image_path}")

        # ðŸ”ƒ Clear memory
        del image
        torch.cuda.empty_cache()
        gc.collect()

    except Exception as e:
        print(f"âŒ Failed to generate image {idx+1}: {e}")
```
